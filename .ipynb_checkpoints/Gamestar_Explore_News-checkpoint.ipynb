# Analyse der News-Artikel von Gamestar.de

In diesem Dokument finden Sie eine Analyse zu den News-Artikeln, die von [Gamestar](http://www.gamestar.de/) veröffentlicht wurden. Dabei werden 22356 Artikel aus den letzen 3 Jahren betrachtet.

# Welche Daten stehen für die Auswertung zur Verfügung?

Für die gesammelten Artikel stehen folgende Werte/Variablen für die Auswertung zur Verfügung:

* **Datum+Uhrzeit:**&nbsp;&nbsp;&nbsp;Datum und Uhrzeit der Veröffentlichung des Artikels
* **Titel:**&nbsp;&nbsp;&nbsp;Titel/Überschrift für den Artikel
* **Thema:**&nbsp;&nbsp;&nbsp;Die Themen werden den ersten Wörtern des Titels entnommen
* **Kommentare:**&nbsp;&nbsp;&nbsp;Anzahl der Kommentare, die der Artikel erhielt

# Wie wurden die Daten gesammelt?

Gesammelt wurden die Daten der einzelnen Tests mit der Python-Library [requests](http://docs.python-requests.org/en/master/). 

Der Programmcode zum Webscraping ist in der Datei "Gamestar_Scrape_News.ipynb" einzusehen.

# Wie wurden die Tests analysiert?

Zum Nachvollziehen der Analyse ist der verwendete Programmcode ausführlich kommentiert 

```python
# Imprort aller benötigten Libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import time
```

```python
# Laden der Daten aus der csv-Datei
df = pd.read_csv('build/Gamestar_News.csv', index_col='Date', parse_dates=True)
print('Daten zu {} Artikeln wurden geladen'.format(df.shape[0]))
```

Jede Reihe stellt einen Artikel dar und die Spalten enthalten die jeweiligen Werte/Variablen der Artikel.

```python
# Beispielhafte Darstellung der Datenstruktur der ersten 5 Artikel
df.head()
```

# Wie entwickelte sich die Anzahl der Artikel über die Zeit?
Die Daten werden monatlich zusammengefasst und die Anzahl der Artikel pro Monat wird dargestellt.

```python
df.groupby(pd.TimeGrouper(freq='m'))['Title'].count().plot()
plt.title('Anzahl der monatlich veröffentlichten Artikel')
plt.tight_layout()
plt.savefig('build/Gamestar_News_Monatliche_Artikel.pdf', bbox_inches='tight')
plt.show()
```

Sichtbar ist, dass zu Beginn des Jahres 2015 eine deutliche Steigerung der Artikelanzahl stattgefunden hat. Über das Jahr 2016 blieb die Artikel-Anzahl relativ konstant und nahm dann über das Jahr 2017 wieder leicht ab.

# Wie entwickelte sich die durchschnittliche Anzahl an Kommentaren zu den Artikeln?
Die Daten werden monatlich zusammengefasst und dargestellt. Dabei wird die Anzahl der erhaltenen Kommentare pro Monat durch die Anzahl der Artikel pro Monat geteilt, sodass die durchschnittliche Anzahl an Kommentaren pro Artikel dargestellt wird.

```python
df.groupby(pd.TimeGrouper(freq='m'))['Comments'].mean().plot()
plt.title('Durchschnittliche Anzahl der Kommentare pro Artikel')
plt.ylim(0, 50)
plt.tight_layout()
plt.savefig('build/Gamestar_News_Monatliche_Kommentare_pro_Artikel.pdf', bbox_inches='tight')
plt.show()
```

* Über das Jahr 2015 nimmt die Zahl der Kommentare pro Artikel stark ab. Dies kann der steigenden Anzahl an Artikeln geschuldet sein, sodass sich die Kommentare nur auf mehr Artikel verteilen.
* Über das Jahr 2016/2017 sinkt die Anzahl an Kommentaren leicht, obwohl gleichbleibend oder sogar weniger Artikel veröffentlicht werden. Dies kann auf weniger interaktionsfreudige Leser oder weniger Leser hindeuten.

# Welche Artikel wurden am häufigsten kommentiert?
Die 10 Artikel mit den meisten Kommentaren sind abgebildet mit ihren jeweiligen Daten.

```python
df.sort_values('Comments', ascending=False).head(10)
```

# Welche Themen haben insgesamt die meisten Kommentare gesammelt?
Gruppiert werden die Artikel hier nach dem ersten Abschnitt ihres Titels. Ein Artikel á la "Overwatch - Halloween-Skins und der neue Horde-Modus im Trailer" hätte den Abschnitt vor dem Bindestrich als Thema ("Overwatch"). Alle Artikel sind nach diesem Thema zusammengefasst und ihre Kommentare werden zusammengezählt. Die 20 meist kommentierten Themen werden dargestellt.

```python
n = 20
plt.figure(figsize=(10, n/4))
sns.barplot(data=df.groupby('Topic').sum().sort_values('Comments', ascending=False).head(n).reset_index(), x='Comments', y='Topic')
plt.xlabel('Kommentare')
plt.tight_layout()
plt.savefig('build/Gamestar_News_Top_Kommentierte_Themen.pdf', bbox_inches='tight')
plt.show()
```

Anzumerken ist, dass die Anzahl der Artikel hier noch nicht eingeflossen ist.

# Welche Themen sammeln überdurchschnittlich viele Kommentare pro Artikel?
Die Artikel werden erneut wie oben nach ihren Themen gruppiert und die Kommentare zusammengezählt. Anschließend wird durch die Anzahl der Artikel geteilt, sodass ein Durchschnitt an Kommentaren pro Thema entsteht.
Es wurden nur Themen mit mindestens 10 Artikeln berücksichtigt.

```python
n = 20
m = 10

df_comments = df.groupby('Topic').agg({'Comments':['mean', 'count']})
plt.figure(figsize=(10, n/4))
sns.barplot(data=df_comments[df_comments['Comments']['count']>=m]['Comments'].sort_values('mean', ascending=False).head(n).reset_index(), x='mean', y='Topic')
plt.xlabel('Durchschnittliche Anzahl an Kommentaren')
plt.tight_layout()
plt.savefig('build/Gamestar_News_Top_Kommentierte_Themen_pro_Artikel.pdf', bbox_inches='tight')
plt.show()
```

"Star Citizen" hält sowohl bei der Anzahl der meisten Kommentare pro Artikel wie auch bei den meisten Kommentaren generell die Spitze. Überraschend ist, dass nur die bekanntesten Spiele viele Kommentare sammeln können und ansonsten Hardware und übergeordnete Themen die Liste dominieren. Auch Nintendo stellt PC-nähere Konsolen hier deutlich in den Schatten.

# Wie entwickelten sich die meist kommentierten Themen über die Zeit?
Die 5 Themen mit den meisten Kommentaren generell werden hier im Zeitverlauf dargestellt. Dafür werden die Kommentare monatlich gruppiert.

```python
n = 10
best_topics = df.groupby('Topic').sum().sort_values('Comments', ascending=False).head(5).index
df[df['Topic'].isin(best_topics)].groupby([pd.TimeGrouper(freq='m'), 'Topic']).agg({'Comments':'sum'}).reset_index().pivot(index='Date', columns='Topic')['Comments'].plot()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Verlauf der {} meist kommentierten Themen'.format(n))
plt.ylabel('Monatliche Kommentare')
plt.tight_layout()
plt.savefig('build/Gamestar_News_Verlauf_Monatliche_Top_Artikel.pdf', bbox_inches='tight')
plt.show()
```

Auffällig ist, dass in den vergangenen 12 Monaten keinerlei größere Themen mit Leserbeteiligung aufgetaucht sind. Die aktuellen Diskussionen werden weiterhin von Themen der letzten Jahre dominiert. Da aktuelle Themen von der schieren Fülle an vergangenen Kommentaren überschattet werden könnten, wird in der folgenden Grafik mehr Wert auf das aktuelle Zeitgeschehen gelegt.

# Wie entwickelten sich die meist kommentierten Themen der letzten 3 Monate über die Zeit?
Für diese Grafik werden die letzten 3 Monate genommen und daraus die Themen mit den meisten Kommentaren herausgesucht, monatlich gruppiert und über die Zeit dargestellt.

```python
n = 5
best_topics = df.loc[:'2017-07-01'].groupby('Topic').sum().sort_values('Comments', ascending=False).head(n).index
df[df['Topic'].isin(best_topics)].groupby([pd.TimeGrouper(freq='m'), 'Topic']).agg({'Comments':'sum'}).reset_index().pivot(index='Date', columns='Topic')['Comments'].plot()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Verlauf der {} meist kommentierten Themen der letzten 3 Monate'.format(n))
plt.ylabel('Monatliche Kommentare')
plt.tight_layout()
plt.savefig('build/Gamestar_News_Verlauf_Monatliche_Aktuelle_Top_Artikel.pdf', bbox_inches='tight')
plt.show()
```

Auch hier zeigt sich, dass die aktuellen Kommentare stark von Themen der letzten Jahre (Star Citizen, Overwatch) oder PC-fremden Themen (Game of Thrones) beherrscht wird.
Klar anzumerken ist, dass dies nur als Momentaufnahme gewertet werden kann und keinerlei Vergleiche mit Vergangenem vorliegen. Sollte ähnliches Verhalten schon immer so gewesen sein, wäre dies keine Besonderheit mehr.

# Wie werden die Artikel über den Tag veröffentlicht?
Mit der Zeit der Veröffentlichung des Artikels lässt sich die redaktionelle Aktivität abschätzen.

```python
dict_week = {0:'Montag', 1:'Dienstag', 2:'Mittwoch', 3:'Donnerstag', 4:'Freitag', 5:'Samstag', 6:'Sonntag'}
_, day = zip(*dict_week.items())

df_week = df
df_week['Weekday'] = df_week.index.weekday
df_week['Weekday'] = df_week['Weekday'].apply(lambda x: dict_week[x])
df_week = df_week.drop(['Title', 'Comments', 'Topic'], axis=1)

df_week.index = df_week.index.time
df_week = df_week.reset_index()

df_pivot = df_week.groupby(['index', 'Weekday'])['Weekday'].count()
df_pivot = df_pivot.rename('Value').reset_index()
df_pivot = df_pivot.pivot(index='index', columns='Weekday')['Value'].reindex_axis(day, axis=1)

df_pivot.fillna(0).rolling(window=60, center=True).mean().plot()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Wann werden die Artikel über den Tag veröffentlicht?')
plt.xlabel('Uhrzeit')
plt.ylabel('Anzahl der Artikel pro Stunde')
plt.tight_layout()
plt.savefig('build/Gamestar_News_Artikel_Time.pdf', bbox_inches='tight')
plt.show()
```

Da die Anzahl der Artikel über 60min gemittelt wurden, bewegen sich die angegebenen Zeiten ebenfalls grob in einem solchen Zeitfenster.
* Am Wochenende kommen weniger Artikel als an Arbeitstagen.
* Die Startzeit in der Redaktion spiegelt sich schön wieder.
* Auch eine Mittagspause lässt sich erahnen.
* Wie vermutlich in jedem Unternehmen wird auch Freitags der Feierabend etwas früher eingeläutet als an den anderen Tagen

# An welchen Wochentagen kommen die meisten Artikel heraus?
Gruppiert wurden die Artikel hier nach ihrem Wochentag.

```python
df.groupby('Weekday').agg({'Title':'count'}).reindex(index=day).plot.barh(legend=None)
plt.title('Artikel aufgeteilt auf die Wochentage')
plt.ylabel('')
plt.xlabel('Anzahl der Artikel')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('build/Gamestar_News_Artikel_Weekday.pdf', bbox_inches='tight')
plt.show()
```

Am produktivsten scheint die Wochenmitte zu sein.

# In welchen Monaten kommen die meisten Artikel heraus?
Gruppiert werden die Artikel nach Monat und die einzelnen Monate werden über die verschiedenen Jahre verglichen.

```python
months = ['Januar', 'Februar', 'März', 'April', 'Mai', 'Juni', 'Juli', 'August', 'September', 'Oktober', 'November', 'Dezember']
dict_month = {i:x for i, x in enumerate(months, 1)}

df['Year'] = df.index.year
df['Month'] = df.index.month
df['Month'] = df['Month'].apply(lambda x: dict_month[x])
```

```python
df.groupby(['Month', 'Year']).agg({'Title':'count'}).reset_index().pivot(index='Month', columns='Year', values='Title').fillna(0).reindex(index=months).plot.barh()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.gca().invert_yaxis()

plt.tight_layout()
plt.savefig('build/Gamestar_News_Artikel_Month.pdf', bbox_inches='tight')
plt.show()
```

Viele der Monate werden von den Jahren 2015/2016 dominiert. Dies lies sich schon in den ersten Grafiken erahnen.

# Wieviele Artikel kommen durchschnittlich pro Tag heraus?
Die Anzahl der Artikel wird auf die Anzahl der Tage im jeweiligen Jahr umgerechnet. Dass das aktuelle Jahr noch andauert ist mit eingerechnet.

```python
def articleYear(year):
    df_temp = df[df['Year']==year.values[0]]
    articles = df_temp.shape[0]
    first = df_temp.index.min()
    last = df_temp.index.max()
    return articles/(last-first).days

df.groupby('Year').agg({'Year':articleYear}).iloc[1:].plot.barh(legend=None)
plt.gca().invert_yaxis()
plt.title('Anzahl der Artikel pro Tag')
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('build/Gamestar_News_Artikel_Year.pdf', bbox_inches='tight')
plt.show()
```
