# Analyse der Spiele-Tests von Gamestar.de

In diesem Dokument finden Sie eine Analyse zu den Computerspiele-Tests, die von [Gamestar](http://www.gamestar.de/) veröffentlicht wurden. Dabei werden 1368 Tests aus den letzen 20 Jahren betrachtet.

# Welche Daten stehen für die Auswertung zur Verfügung?

Für die gesammelten Tests stehen folgende Werte/Variablen für die Auswertung zur Verfügung:

* **Wertung:**&nbsp;&nbsp;&nbsp;Wertung des Spieles
* **Name:**&nbsp;&nbsp;&nbsp;Name des Spiels
* **Link:**&nbsp;&nbsp;&nbsp;Link zum Test
* **Tag:**&nbsp;&nbsp;&nbsp;Tag der Veröffentlichung des Spiels
* **Monat:**&nbsp;&nbsp;&nbsp;Monat der Veröffentlichung des Spiels
* **Jahr:**&nbsp;&nbsp;&nbsp;Jahr der Veröffentlichung des Spiels
* **Quartal:**&nbsp;&nbsp;&nbsp;Quartal der Veröffentlichung des Spiels

* **Kategorie:**&nbsp;&nbsp;&nbsp;Kategorie für das Spiel
* **Publisher:**&nbsp;&nbsp;&nbsp;Publisher des Spieles

# Wie wurden die Daten gesammelt?

Gesammelt wurden die Daten der einzelnen Tests mit der Python-Library [requests](http://docs.python-requests.org/en/master/). 
* Zuerst wurden die Links, Namen, Release-Daten und Wertungen aller Spieletests auf [dieser und folgenden Unterseiten](http://www.gamestar.de/tests/) gesammelt
* Anschließend wurden zu allen Spielen die Wertungsseite (z.B. [Raiders of the Broken Planet](http://www.gamestar.de/spiele/raiders-of-the-broken-planet/wertung/53750.html)) aufgreufen und aus der Kopfzeile die Kategorie und der Publisher entnommen

Der Programmcode zum Webscraping ist in der Datei "Gamestar_Scrape_Tests-ipynb" einzusehen.

# Wie sind die Werte für die einzelnen Variablen verteilt?
Diese Darstellungen entsprechen in der allerersten Matrixübersicht den Diagrammen [01], [13] und [19]. Jeder Balken ist hierbei einem Wert auf der unteren x-Achse zugeordnet und die Höhe entspricht der Anzahl der Tests die diesen x-Wert aufweisen. Unterhalb der Diagramme sind die Auffälligkeiten der Diagramme beschrieben.

```python
def plotHistogram(df, col, bin_width=1):
    bin_min = int(df[col].min())
    bin_max = int(df[col].max())
    df[col].hist(bins=range(bin_min, bin_max+1+bin_width, bin_width), color=rgb)
    plt.title('Verteilung der Werte in der Variable {}'.format(col))
    plt.xlabel(col)
    plt.ylabel('Anzahl')
    plt.tight_layout()
    plt.savefig('build/Gamestar_Tests_Histogram_{}.pdf'.format(col), bbox_inches='tight')
    plt.show()
    
    
for col in ['Wertung', 'Monat', 'Jahr']:
    plotHistogram(df, col)
```

Diagramm 1 - [01]:
* Die meisten Bewertungen sind knapp über 80 Punkten anzutreffen.
* Genau 80 und 85 Punkte werden besonders häufig vergeben.
* Wie 91 oder mehr ist auch 49 oder weniger ein seltenes Ereignis

Diagramm 2 - [13]:
* Die Release-Daten liegen meistens vor den Sommer- und den Weihnachtsferien.
* In den Monaten vor Weihnachten liegen am meisten Spieleveröffentlichungen.

Diagramm 3 - [19]:
* Tests für Spiele sind online erst ab dem Jahr 2000 abrufbar (mit einer Ausnahme).
* 2007 wurden deutlich weniger und 2009 deutlich mehr Spiele getestet als erwartet.

# Veränderte sich die Bewertung über die Jahre?
Die einzelnen Spielebewertungen werden hier nach ihrem Jahr der Veröffentlichung sortiert dargestellt. Hinzugefügt wird eine Trendlinie. Diese Trendlinie ist gerade an ihren beiden Enden mit Vorsicht zu betrachten, da wenige Daten dort vorliegen und der Trend nur geringfügig für die Zukunft gelten kann.

```python
for elem in ['Wertung']:
    x = df['Jahr']
    y = df[elem]
    
    sns.regplot(x, y, lowess=True, color='firebrick', scatter_kws={'alpha':0.5, 'color':rgb})
    plt.title('Verteilung der Bewertungen über die Jahre')
    plt.tight_layout()
    plt.savefig('build/Gamestar_Tests_{}.pdf'.format(elem), bbox_inches='tight')
    plt.show()
```

Zu sehen ist, dass vor 2000 nur ein einzelnes Spiel online abrufbar war. Auch wird deutlich, dass in den ersten Jahren nach der Jahrtausendwende Spiele durchschnittlich eine höhere Bewertung hatten als um etwa 2010. Seit diesem Jahr nimmt die durchschnittliche Punktzahl pro Spiel wieder zu.

An dieser Grafik lässt sich sehr schön das zweischneidige Schwert jeder Datenanalyse zeigen. Was ist die Ursache der Trendlinie?
* War die Gamestar-Redaktion um 2010 pessimistisch drauf und bewertete die Spiele deshalb schlecht? (Ursache liegt in der Wertung/Redaktion)
* Wurden um 2010 nur schlechte Spiele produziert und die Spiele verdienen weniger Punkte? (Ursache liegt im Jahr/Publisher)
* Oder hängt alles von einem dritten noch unbekannten Faktor ab? (Wetter, Finanzkrise, ...)

Woran diese Entwicklung liegt, lässt sich nur schwerlich bestimmen und bedarf menschlicher Interpretation, die wiederum vielfach ausgelegt werden kann. Schlüsse aus Grafiken zu ziehen bedarf Vorsicht.

# Welche Kategorien enthalten gute Spiele?
Für dieses Diagramm werden die einzelnen Tests nach ihrer Kategorie zusammen gefasst. Dabei werden nur Kategorien mit mindestens 10 Spielen berücksichtigt um Ausreißer in den Daten zu minimieren.
* Jeder Punkt im Diagramm stellt ein Spiel dar.
* Sortiert sind die Kategorien nach ihrer durchschnittlichen Wertung.
* Die Farbe ist bedeutungslos.

```python
def boxplot_sorted(df, by, column, n=5, x=10, y=10):
    plt.figure(figsize=(x,y))
    s = df[by].value_counts()>=n
    s = s[s.values].index
    
    df = pd.DataFrame({col:vals[column] for col, vals in df.groupby([by]) if col in s})
    meds = df.median().sort_values()
    df = pd.melt(df[meds.index]).dropna()
    df.columns = [by, column]
    
    sns.swarmplot(data=df, x=column, y=by)
    plt.title('{} mit mindestens {} Daten verglichen mit ihrer {}'.format(by, n, column))
    plt.xlabel(column)
    plt.ylabel(by)
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig('build/Gamestar_Tests_Swarmplot_{}_{}.pdf'.format(by, column), bbox_inches='tight')
    plt.show()
```

```python
boxplot_sorted(df, by='Kategorie', column='Wertung', n=10, x=10, y=18)
```

Sport-Simulationen scheinen die höchsten Bewertungen zu erhalten, während populärere (häufig getestetere) Kategorien eher auch mal schwarze Schafe unter sich verstecken.

# Interaktive Umsetzung für den Publisher-Wertung Vergleich
Dieses Diagramm bildet das vorangehende interaktiv nach. Es kann im Ordner 'build' unter dem Dateinamen 'Gamestar_Tests_Publisher_Interaktiv.html' gefunden werden. Die Interaktion ermöglicht es die einzelnen Punkte mit der Maus auf Ihren Namen, ihre Wertung und weitere Informationen zu untersuchen. Beim Klicken auf einzelne Punkte, werden Sie auf die Spiele-Test-Seite weitergeleitet.

```python
s = df['Publisher'].value_counts()>=20
s = s[s.values].index
    
df_plot = df[df['Publisher'].isin(s)][['Wertung', 'Name', 'Link', 'Kategorie', 'Publisher', 'Jahr']]


dict_pos = {name:i for i, name in enumerate(s)}
inv_pos = {v: k for k, v in dict_pos.items()}

def swarmplotTransform(x):
    return dict_pos[x]+np.diff(np.random.random(2))[0]/3

df_plot['Publisher_pos'] = df_plot['Publisher'].apply(lambda x: swarmplotTransform(x))
df_plot['Wertung_pos'] = df_plot['Wertung'].apply(lambda x: x+np.diff(np.random.random(2))[0]/3)

import matplotlib.cm as cm
import matplotlib

norm = matplotlib.colors.Normalize(vmin=0, vmax=len(dict_pos)*3, clip=True)
mapper = cm.ScalarMappable(norm=norm, cmap=cm.prism)#cm.nipy_spectral)

def rgba_to_hex(color):
    '''
    Needs a rgba-color tuple
        
    Returns the converted hex-code color
    '''
    return '#%02x%02x%02x' % (int(color[0]*255), int(color[1]*255), int(color[2]*255))

df_plot['Color'] = df_plot['Publisher'].apply(lambda x: dict_pos[x])
df_plot['Color'] = df_plot['Color'].apply(mapper.to_rgba).apply(rgba_to_hex)
```

```python
from bokeh.plotting import figure, show, output_file
from bokeh.models import FuncTickFormatter, OpenURL, ColumnDataSource, HoverTool, TapTool, Range1d, PanTool, ResetTool, SaveTool, WheelZoomTool
from bokeh.models.tickers import FixedTicker

title = 'Publisher mit mindestens 20 Spielen verglichen mit ihrer Wertung'

# Load useful tools
pan = PanTool()
reset = ResetTool()
save = SaveTool()
zoom = WheelZoomTool()
tap = TapTool()

# Create the hovertool with its data
hover = HoverTool(tooltips=[
    ("Name", "@Name"),
    ("Wertung", "@Wertung"),
    ("Kategorie", "@Kategorie"),
    ("Publisher", "@Publisher"),
    ("Jahr", "@Jahr")
])

p = figure(plot_width=600, plot_height=300, responsive = True, tools=[hover, pan, reset, save, zoom, tap], active_scroll=zoom, title=title, toolbar_location="above")
p.xaxis.axis_label = 'Wertung'
p.yaxis.axis_label = 'Publisher'

source = ColumnDataSource(df_plot)

p.circle(source=source, x='Wertung_pos', y='Publisher_pos', color='Color', size=6)
p.yaxis.ticker = FixedTicker(ticks=list(range(len(inv_pos))))

p.yaxis.formatter = FuncTickFormatter(code="""
    var labels = %s;
    return labels[tick];
""" % inv_pos)

# Add an interactive taptool
taptool = p.select(type=TapTool)
taptool.callback = OpenURL(url='@Link')

p.xgrid.bounds = (0, 100)
p.ygrid.bounds = (0, len(dict_pos))

p.y_range = Range1d(len(inv_pos), -1)


output_file('build/Gamestar_Tests_Publisher_Interaktiv.html', title='Gamestar Tests')
show(p)
```

# Wann veröffentlichen die Publisher ihre Spiele und wie hängt die Qualität vom Monat ab?
Gruppiert werden die Spiele-Tests für diese Grafik nach dem Monat in dem sie erschienen sind. Muster über den Jahresverlauf können somit dargestellt werden. Um die Lesbarkeit des Diagramms zu gewährleisten, sind nur die 3 Publisher mit den meisten Spielen integriert.

```python
df['Datum'] = df['Jahr'].fillna(0).astype(np.int64).astype(str)+'-'+df['Monat'].fillna(0).astype(np.int64).astype(str)+'-'+df['Tag'].fillna(0).astype(np.int64).astype(str)
df['Datum'] = pd.to_datetime(df['Datum'], errors='coerce')
df = df.set_index('Datum', drop=True)
```

```python
def plotTimeseries(n, column, group1, group2, title, ylabel, file_extension, typ='count'):
    months = list(range(int(df[group1].min()), int(df[group1].max()+1)))
    most_contributing = df.groupby(group2)['Name'].count().sort_values(ascending=False)
    if typ=='count':
        df_plot = df.groupby([group1, group2])[column].count().unstack(level=-1).reindex(index=months)
    else:
        df_plot = df.groupby([group1, group2])[column].mean().unstack(level=-1).reindex(index=months)

    df_plot[most_contributing.index[:n]].plot()
    plt.title(title.format(n, group2))
    plt.ylabel(ylabel)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    plt.tight_layout()
    plt.savefig('build/Gamestar_Tests_{}_{}_{}.pdf'.format(file_extension, group2, group1), bbox_inches='tight')
    plt.show()
```

```python
n = 3
column = 'Name'
group1 = 'Monat'
group2 = 'Publisher'
title = 'Veröffentlichungen der {} fleißigsten {}'
ylabel = 'Anzahl veröffentlichter Spiele'
file_extension = 'Anzahl'

plotTimeseries(n, column, group1, group2, title, ylabel, file_extension)


n = 3
column = 'Wertung'
group1 = 'Monat'
group2 = 'Publisher'
title = 'Veröffentlichungen der {} fleißigsten {}'
ylabel = 'Wertung der Spiele'
file_extension = 'Wertung'

plotTimeseries(n, column, group1, group2, title, ylabel, file_extension, typ='mean')
```

Diagramm 1 - Wann veröffentlichen die Publisher über das Jahr gesehen ihre Spiele?:
* Deutlich tritt hervor, dass Electronic Arts seine Spiele häufig vor Weihnachten veröffentlicht. 
* Da Daten aus knapp 20 Jahren einfließen, ist es verwunderlich, dass euch die Top Publisher Monate ohne eine einzige Spieleveröffentlichung aufweisen.

Diagramm 2 - Welche Bewertungen kriegen die Spiele über das Jahr gesehen?:
* Ubisoft kann auf das gesamte Jahr die stabilsten Werte für seine Spiele vorweisen.
* Deep Silver und Electronic Arts scheinen in der Jahresmitte beide ein Qualitätstief aufzuweisen. (Dies kann an der geringen Anzahl der Spiele in der Sommerpause liegen, sodass ein schlechtes Spiel in der Pause die wenigen guten stark mit beeinflusst)

```python
from math import pi
import pandas as pd

from bokeh.io import show
from bokeh.models import ColumnDataSource, HoverTool, LinearColorMapper, BasicTicker, PrintfTickFormatter, ColorBar
from bokeh.plotting import figure
from bokeh.palettes import Plasma256

mapper = LinearColorMapper(palette=Plasma256, low=df_plot['Count'].min(), high=df_plot['Count'].max())
source = ColumnDataSource(df_plot)

TOOLS = "hover,save"

p = figure(x_axis_location="above", plot_width=800, plot_height=1600,
           tools=TOOLS, toolbar_location='left', responsive=True, title='Welche Kategorien bedienen die Publisher?')

p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "8pt"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = pi / 3

p.rect(x="Kategorie_Pos", y="Publisher_Pos", width=1, height=1,
       source=source,
       fill_color={'field': 'Count', 'transform': mapper},
       line_color=None)

color_bar = ColorBar(color_mapper=mapper, major_label_text_font_size="8pt",
                     ticker=BasicTicker(desired_num_ticks=20),
                     label_standoff=6, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

p.select_one(HoverTool).tooltips = [
     ('Publisher', '@Publisher'),
     ('Kategorie', '@Kategorie'),
     ('Anzahl', '@Count')]

p.yaxis.ticker = FixedTicker(ticks=list(range(len(dict_publisher))))
p.yaxis.formatter = FuncTickFormatter(code="""
    var labels = %s;
    return labels[tick];
""" % dict_publisher)

p.xaxis.ticker = FixedTicker(ticks=list(range(len(dict_kategorie))))
p.xaxis.formatter = FuncTickFormatter(code="""
    var labels = %s;
    return labels[tick];
""" % dict_kategorie)

output_file('build/Gamestar_Tests_Publisher_Kategorie_Interaktiv.html', title='Gamestar Tests')
show(p)
```
