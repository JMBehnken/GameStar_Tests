# Auswertung der Spiele-Tests von Gamestar.de

Für die gesammelten Tests stehen folgende Variablen für die Auswertung zur Verfügung:

* **Wertung:**&nbsp;&nbsp;&nbsp;Wertung des Spieles
* **Name:**&nbsp;&nbsp;&nbsp;Name des Spiels
* **Link:**&nbsp;&nbsp;&nbsp;Link zum Test
* **Tag:**&nbsp;&nbsp;&nbsp;Tag der Veröffentlichung des Spiels
* **Monat:**&nbsp;&nbsp;&nbsp;Monat der Veröffentlichung des Spiels
* **Jahr:**&nbsp;&nbsp;&nbsp;Jahr der Veröffentlichung des Spiels
* **Quartal:**&nbsp;&nbsp;&nbsp;Quartal der Veröffentlichung des Spiels
* **Plattform:**&nbsp;&nbsp;&nbsp;Plattform für die der Test durchgeführt wurde
* **Kategorie:**&nbsp;&nbsp;&nbsp;Kategorie für das Spiel
* **Publisher:**&nbsp;&nbsp;&nbsp;Publisher des Spieles

```python
# Import
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Loading the data
df = pd.read_csv('build/Gamestar_Test_Combined.csv')

dict_month = {np.nan:None, 
              'Januar':1, 
              'Februar':2, 
              'März':3, 
              'April':4, 
              'Mai':5, 
              'Juni':6, 
              'Juli':7, 
              'August':8, 
              'September':9, 
              'Oktober':10, 
              'November':11, 
              'Dezember':12}

# Transforming the months
df['Monat'] = df['Monat'].apply(lambda x: dict_month[x])
```

# Wie sind die Werte für die einzelnen Variablen verteilt?

```python
def plotHistogram(df, col, bin_width=1):
    bin_min = int(df[col].min())
    bin_max = int(df[col].max())
    df[col].hist(bins=range(bin_min, bin_max+1+bin_width, bin_width), color=rgb)
    plt.title('Verteilung der Werte in der Variable {}'.format(col))
    plt.xlabel(col)
    plt.ylabel('Anzahl')
    plt.tight_layout()
    plt.savefig('build/Gamestar_Tests_Histogram_{}.pdf'.format(col))
    plt.show()
    
    
for col in ['Wertung', 'Monat', 'Jahr']:
    plotHistogram(df, col)
```

# Veränderte sich die Bewertung über die Jahre?

```python
for elem in ['Wertung']:
    x = df['Jahr']
    y = df[elem]
    
    sns.regplot(x, y, lowess=True, color='firebrick', scatter_kws={'alpha':0.5, 'color':rgb})
    plt.title('Verteilung der Bewertungen über die Jahre')
    plt.tight_layout()
    plt.savefig('build/Gamestar_Tests_{}.pdf'.format(elem))
    plt.show()
```

# Wie häufig wurden einzelne Kategorien getestet?

# Welche Kategorien enthalten gute Spiele?

```python
def boxplot_sorted(df, by, column, n=5, x=10, y=10):
    plt.figure(figsize=(x,y))
    s = df[by].value_counts()>=n
    s = s[s.values].index
    
    df = pd.DataFrame({col:vals[column] for col, vals in df.groupby([by]) if col in s})
    meds = df.median().sort_values()
    df = pd.melt(df[meds.index]).dropna()
    df.columns = [by, column]
    
    sns.swarmplot(data=df, x=column, y=by)
    plt.title('{} mit mindestens {} Daten verglichen mit ihrer {}'.format(by, n, column))
    plt.xlabel(column)
    plt.ylabel(by)
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig('build/Gamestar_Tests_Swarmplot_{}_{}.pdf'.format(by, column))
    plt.show()
```

```python
boxplot_sorted(df, by='Kategorie', column='Wertung', n=10, x=10, y=18)
```

# Welche Publisher produzieren gute Spiele?

```python
boxplot_sorted(df, by='Publisher', column='Wertung', n=20, x=10, y=12)
```

# Interaktive Umsetzung für den Publisher-Wertung Vergleich

```python
s = df['Publisher'].value_counts()>=20
s = s[s.values].index
    
df_plot = df[df['Publisher'].isin(s)][['Wertung', 'Name', 'Link', 'Kategorie', 'Publisher', 'Jahr']]


dict_pos = {name:i for i, name in enumerate(s)}
inv_pos = {v: k for k, v in dict_pos.items()}

def swarmplotTransform(x):
    return dict_pos[x]+np.diff(np.random.random(2))[0]/3

df_plot['Publisher_pos'] = df_plot['Publisher'].apply(lambda x: swarmplotTransform(x))
df_plot['Wertung_pos'] = df_plot['Wertung'].apply(lambda x: x+np.diff(np.random.random(2))[0]/3)

import matplotlib.cm as cm
import matplotlib

norm = matplotlib.colors.Normalize(vmin=0, vmax=len(dict_pos)*3, clip=True)
mapper = cm.ScalarMappable(norm=norm, cmap=cm.prism)#cm.nipy_spectral)

def rgba_to_hex(color):
    '''
    Needs a rgba-color tuple
        
    Returns the converted hex-code color
    '''
    return '#%02x%02x%02x' % (int(color[0]*255), int(color[1]*255), int(color[2]*255))

df_plot['Color'] = df_plot['Publisher'].apply(lambda x: dict_pos[x])
df_plot['Color'] = df_plot['Color'].apply(mapper.to_rgba).apply(rgba_to_hex)
```

```python
from bokeh.plotting import figure, show, output_file
from bokeh.models import FuncTickFormatter, OpenURL, ColumnDataSource, HoverTool, TapTool, Range1d, PanTool, ResetTool, SaveTool, WheelZoomTool
from bokeh.models.tickers import FixedTicker

title = 'Publisher mit mindestens 20 Spielen verglichen mit ihrer Wertung'

# Load useful tools
pan = PanTool()
reset = ResetTool()
save = SaveTool()
zoom = WheelZoomTool()
tap = TapTool()

# Create the hovertool with its data
hover = HoverTool(tooltips=[
    ("Name", "@Name"),
    ("Wertung", "@Wertung"),
    ("Kategorie", "@Kategorie"),
    ("Publisher", "@Publisher"),
    ("Jahr", "@Jahr")
])

p = figure(plot_width=600, plot_height=300, responsive = True, tools=[hover, pan, reset, save, zoom, tap], active_scroll=zoom, title=title, toolbar_location="above")
p.xaxis.axis_label = 'Wertung'
p.yaxis.axis_label = 'Publisher'

source = ColumnDataSource(df_plot)

p.circle(source=source, x='Wertung_pos', y='Publisher_pos', color='Color', size=6)
p.yaxis.ticker = FixedTicker(ticks=list(range(len(inv_pos))))

p.yaxis.formatter = FuncTickFormatter(code="""
    var labels = %s;
    return labels[tick];
""" % inv_pos)

# Add an interactive taptool
taptool = p.select(type=TapTool)
taptool.callback = OpenURL(url='@Link')

p.xgrid.bounds = (0, 100)
p.ygrid.bounds = (0, len(dict_pos))

p.y_range = Range1d(len(inv_pos), -1)


output_file('build/Gamestar_Tests_Publisher_Interaktive.html', title='Gamestar Tests')
show(p)
```
